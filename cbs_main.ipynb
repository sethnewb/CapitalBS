```python
# url to capital bike share api:
# http://opendata.dc.gov/datasets/capital-bike-share-locations/geoservice
# api query url:
# https://maps2.dcgis.dc.gov/dcgis/rest/services/DCGIS_DATA/Transportation_WebMercator/MapServer/5/query?where=1%3D1&outFields=*&outSR=4326&f=json

import pymongo
import requests
import pandas as pd
from pprint import pprint as p
import datetime as dt


# conn = 'mongodb://localhost:27017'
# client = pymongo.MongoClient(conn)
# db = client.bikeshare

url = "https://maps2.dcgis.dc.gov/dcgis/rest/services/DCGIS_DATA/Transportation_WebMercator/MapServer/5/query?where=1%3D1&outFields=*&outSR=4326&f=json"
response = requests.get(url).json()

data = response["features"]

p(data)
```
